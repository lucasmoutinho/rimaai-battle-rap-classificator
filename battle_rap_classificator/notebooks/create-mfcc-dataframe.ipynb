```python
import warnings
import sklearn

def mfcc_extract(url):
    audio_path = '../resources/audio_samples/soundfile.aac'
    r = requests.get(url, allow_redirects=True)
    open(audio_path, 'wb').write(r.content)
    warnings.filterwarnings("ignore")
    x, sr = librosa.load(audio_path)
    mfccs = librosa.feature.mfcc(x, sr=sr)

    # Feature scaling
    mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
    
    mfccs_processed = np.mean(mfccs.T,axis=0)
    return mfccs_processed
```

```python
# Testing concatenation
a = mfcc_extract(dataset["Ataque"][0])
b = mfcc_extract(dataset["Resposta"][0])
c = np.concatenate([a,b])
print(c)
```

```python
features = []
idx = 0
for a_url, r_url, result in zip(dataset["Ataque"], dataset["Resposta"], dataset["Resultado"]):
    features.append([np.concatenate([mfcc_extract(a_url), mfcc_extract(r_url)]), result])
    if(idx % 100 == 0):
        print(idx)
    idx = idx + 1
```

```python
df = pd.DataFrame(features, columns=['feature','class_label'])
```
